```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from sklearn.metrics import roc_auc_score, average_precision_score
import matplotlib.pyplot as plt

# ----------------------------
# 1. Synthetic structured data
# ----------------------------
num_orders = 50
num_products = 20
num_parts = 30
emb_dim = 32
num_heads = 4

# Optional: synthetic order metadata (e.g., 5 features per order)
order_metadata = torch.randn(num_orders, 5)

# Each product has a subset of parts
product_to_parts = {i: torch.randint(0, num_parts, (5,)) for i in range(num_products)}

# Each order contains a subset of products
orders_to_products = {i: torch.randint(0, num_products, (3,)) for i in range(num_orders)}

# Build order->part matrix
R = torch.zeros((num_orders, num_parts))
for o, products in orders_to_products.items():
    parts = torch.cat([product_to_parts[int(p)] for p in products])
    R[o, parts] = 1

# Train/test split of positive edges
pos_edge_index = torch.nonzero(R).T
perm = torch.randperm(pos_edge_index.size(1))
train_idx = perm[:int(0.8*pos_edge_index.size(1))]
test_idx = perm[int(0.8*pos_edge_index.size(1)):]
train_edges = pos_edge_index[:, train_idx]
test_edges = pos_edge_index[:, test_idx]

# ----------------------------
# 2. Full end-to-end model
# ----------------------------
class FullOrderPartModel(nn.Module):
    def __init__(self, num_orders, num_parts, order_feat_dim=5, emb_dim=32, num_heads=4):
        super().__init__()
        self.order_emb = nn.Embedding(num_orders, emb_dim)
        self.part_emb = nn.Embedding(num_parts, emb_dim)
        self.order_mlp = nn.Sequential(
            nn.Linear(order_feat_dim, emb_dim),
            nn.ReLU(),
            nn.Linear(emb_dim, emb_dim)
        )
        self.phi = nn.Linear(emb_dim, emb_dim)
        self.attention = nn.MultiheadAttention(embed_dim=emb_dim, num_heads=num_heads, batch_first=True)
        self.rho = nn.Sequential(
            nn.Linear(emb_dim*2, emb_dim),
            nn.ReLU(),
            nn.Linear(emb_dim, emb_dim)
        )
        self.decoder = nn.Sequential(
            nn.Linear(emb_dim*2, emb_dim),
            nn.ReLU(),
            nn.Linear(emb_dim, 1)
        )

    def forward(self, order_idx, part_idx, order_features=None):
        order_vec = self.order_emb(order_idx)
        if order_features is not None:
            meta_vec = self.order_mlp(order_features)
            order_vec = torch.cat([order_vec, meta_vec], dim=-1)
            order_vec = self.rho(order_vec)
        part_vec = self.part_emb(part_idx)
        scores = self.decoder(torch.cat([order_vec, part_vec], dim=-1)).squeeze(-1)
        return scores

    def aggregate_new_order(self, part_indices, order_features=None):
        part_embs = self.phi(self.part_emb(part_indices)).unsqueeze(0)
        att_out, _ = self.attention(part_embs, part_embs, part_embs)
        agg_emb = att_out.mean(dim=1)
        if order_features is not None:
            meta_vec = self.order_mlp(order_features)
            agg_emb = self.rho(torch.cat([agg_emb, meta_vec], dim=-1))
        return agg_emb

device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = FullOrderPartModel(num_orders, num_parts, order_feat_dim=5, emb_dim=32, num_heads=4).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
order_metadata = order_metadata.to(device)

# ----------------------------
# 3. Training loop
# ----------------------------
train_losses, val_aucs, val_aps = [], [], []

for epoch in range(50):
    model.train()
    optimizer.zero_grad()

    pos_orders, pos_parts = train_edges.to(device)
    pos_labels = torch.ones(pos_orders.size(0), device=device)

    neg_orders = torch.randint(0, num_orders, (pos_orders.size(0),), device=device)
    neg_parts = torch.randint(0, num_parts, (pos_orders.size(0),), device=device)
    neg_labels = torch.zeros(neg_orders.size(0), device=device)

    all_orders = torch.cat([pos_orders, neg_orders])
    all_parts = torch.cat([pos_parts, neg_parts])
    all_labels = torch.cat([pos_labels, neg_labels])
    all_features = order_metadata[all_orders]

    logits = model(all_orders, all_parts, all_features)
    loss = F.binary_cross_entropy_with_logits(logits, all_labels)
    loss.backward()
    optimizer.step()
    train_losses.append(loss.item())

    # Validation
    model.eval()
    with torch.no_grad():
        test_orders, test_parts = test_edges.to(device)
        neg_test_orders = torch.randint(0, num_orders, (test_orders.size(0),), device=device)
        neg_test_parts = torch.randint(0, num_parts, (test_orders.size(0),), device=device)

        all_test_orders = torch.cat([test_orders, neg_test_orders])
        all_test_parts = torch.cat([test_parts, neg_test_parts])
        all_test_labels = torch.cat([
            torch.ones(test_orders.size(0), device=device),
            torch.zeros(test_orders.size(0), device=device)
        ])
        all_test_features = order_metadata[all_test_orders]

        test_logits = torch.sigmoid(model(all_test_orders, all_test_parts, all_test_features))
        auc = roc_auc_score(all_test_labels.cpu(), test_logits.cpu())
        ap = average_precision_score(all_test_labels.cpu(), test_logits.cpu())
        val_aucs.append(auc)
        val_aps.append(ap)

    if epoch % 5 == 0:
        print(f"Epoch {epoch:02d}, Loss {loss:.4f}, Val AUC {auc:.4f}, AP {ap:.4f}")

# ----------------------------
# 4. Plot training curves
# ----------------------------
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(train_losses); plt.title("Train Loss"); plt.xlabel("Epoch"); plt.grid(True)
plt.subplot(1,3,2)
plt.plot(val_aucs); plt.title("Val ROC-AUC"); plt.xlabel("Epoch"); plt.grid(True)
plt.subplot(1,3,3)
plt.plot(val_aps); plt.title("Val AP"); plt.xlabel("Epoch"); plt.grid(True)
plt.show()

# ----------------------------
# 5. Inference for new order
# ----------------------------
model.eval()
with torch.no_grad():
    new_order_products = torch.tensor([0,5,7], device=device)
    parts_to_aggregate = torch.cat([product_to_parts[int(p)] for p in new_order_products])
    new_order_features = order_metadata[0].unsqueeze(0)  # example metadata
    new_order_emb = model.aggregate_new_order(parts_to_aggregate, new_order_features)

    candidate_parts = torch.arange(num_parts, device=device)
    scores = []
    for p in candidate_parts:
        part_emb = model.part_emb(p).unsqueeze(0)
        score = model.decoder(torch.cat([new_order_emb, part_emb], dim=-1)).item()
        scores.append(score)

    topk = torch.topk(torch.tensor(scores), k=5)
    print("Top predicted parts for new order:", topk.indices.tolist())
```
