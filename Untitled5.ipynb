```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.data import HeteroData
from torch_geometric.nn import SAGEConv, HeteroConv
from torch_geometric.utils import negative_sampling
from sklearn.metrics import roc_auc_score, average_precision_score

# ----------------------------
# 1. Synthetic hetero graph
# ----------------------------
num_orders = 5000
num_products = 200
num_parts = 300

data = HeteroData()
data['order'].x = torch.randn(num_orders, 16)
data['product'].x = torch.randn(num_products, 16)
data['part'].x = torch.randn(num_parts, 16)

# edges: order -> product
data['order', 'buys', 'product'].edge_index = torch.stack([
    torch.randint(0, num_orders, (200,)),
    torch.randint(0, num_products, (200,))
], dim=0)

# edges: product -> part
data['product', 'contains', 'part'].edge_index = torch.stack([
    torch.randint(0, num_products, (300,)),
    torch.randint(0, num_parts, (300,))
], dim=0)

# edges: order -> part (ground-truth)
edge_index = torch.stack([
    torch.randint(0, num_orders, (150,)),
    torch.randint(0, num_parts, (150,))
], dim=0)

# Train/test split
perm = torch.randperm(edge_index.size(1))
train_idx = perm[:int(0.8*edge_index.size(1))]
test_idx = perm[int(0.8*edge_index.size(1)):]
data['order', 'needs', 'part'].edge_index = edge_index[:, train_idx]
test_edges = edge_index[:, test_idx]

# Add reverse edges
data['product', 'rev_buys', 'order'].edge_index = data['order', 'buys', 'product'].edge_index.flip(0)
data['part', 'rev_contains', 'product'].edge_index = data['product', 'contains', 'part'].edge_index.flip(0)
data['part', 'rev_needs', 'order'].edge_index = data['order', 'needs', 'part'].edge_index.flip(0)

# ----------------------------
# 2. HeteroGNN encoder
# ----------------------------
class HeteroGNN(nn.Module):
    def __init__(self, hidden_channels):
        super().__init__()
        self.conv1 = HeteroConv({
            ('order', 'buys', 'product'): SAGEConv((-1, -1), hidden_channels),
            ('product', 'contains', 'part'): SAGEConv((-1, -1), hidden_channels),
            ('order', 'needs', 'part'): SAGEConv((-1, -1), hidden_channels),
            ('product', 'rev_buys', 'order'): SAGEConv((-1, -1), hidden_channels),
            ('part', 'rev_contains', 'product'): SAGEConv((-1, -1), hidden_channels),
            ('part', 'rev_needs', 'order'): SAGEConv((-1, -1), hidden_channels),
        }, aggr='mean')

        self.conv2 = HeteroConv({
            ('order', 'buys', 'product'): SAGEConv((-1, -1), hidden_channels),
            ('product', 'contains', 'part'): SAGEConv((-1, -1), hidden_channels),
            ('order', 'needs', 'part'): SAGEConv((-1, -1), hidden_channels),
            ('product', 'rev_buys', 'order'): SAGEConv((-1, -1), hidden_channels),
            ('part', 'rev_contains', 'product'): SAGEConv((-1, -1), hidden_channels),
            ('part', 'rev_needs', 'order'): SAGEConv((-1, -1), hidden_channels),
        }, aggr='mean')

    def forward(self, x_dict, edge_index_dict):
        x_dict = self.conv1(x_dict, edge_index_dict)
        x_dict = {k: F.relu(v) for k, v in x_dict.items()}
        x_dict = self.conv2(x_dict, edge_index_dict)
        return x_dict

# ----------------------------
# 3. Dot-product decoder
# ----------------------------
class DotProductDecoder(nn.Module):
    def forward(self, z_dict, edge_index):
        src, dst = edge_index
        return (z_dict['order'][src] * z_dict['part'][dst]).sum(dim=-1)

# ----------------------------
# 4. Training helpers
# ----------------------------
def get_link_labels(pos_edge_index, neg_edge_index):
    pos_y = torch.ones(pos_edge_index.size(1))
    neg_y = torch.zeros(neg_edge_index.size(1))
    return torch.cat([pos_y, neg_y], dim=0)

# ----------------------------
# 5. Model + optimizer
# ----------------------------
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = HeteroGNN(hidden_channels=32).to(device)
decoder = DotProductDecoder()
data = data.to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# ----------------------------
# 6. Training loop
# ----------------------------
for epoch in range(20):
    model.train()
    optimizer.zero_grad()
    z_dict = model(data.x_dict, data.edge_index_dict)

    # Negative sampling
    pos_edge_index = data['order', 'needs', 'part'].edge_index
    neg_edge_index = negative_sampling(
        edge_index=pos_edge_index,
        num_nodes=(num_orders, num_parts),
        num_neg_samples=pos_edge_index.size(1)
    )

    edge_label_index = torch.cat([pos_edge_index, neg_edge_index], dim=1)
    edge_label = get_link_labels(pos_edge_index, neg_edge_index).to(device)

    out = decoder(z_dict, edge_label_index)
    loss = F.binary_cross_entropy_with_logits(out, edge_label)
    loss.backward()
    optimizer.step()

    # Evaluate on test edges
    model.eval()
    with torch.no_grad():
        z_dict = model(data.x_dict, data.edge_index_dict)
        neg_test = negative_sampling(
            edge_index=data['order', 'needs', 'part'].edge_index,
            num_nodes=(num_orders, num_parts),
            num_neg_samples=test_edges.size(1)
        )
        edge_label_index_test = torch.cat([test_edges, neg_test], dim=1)
        edge_label_test = get_link_labels(test_edges, neg_test).to(device)
        pred_test = decoder(z_dict, edge_label_index_test).sigmoid()
        auc = roc_auc_score(edge_label_test.cpu(), pred_test.cpu())
        ap = average_precision_score(edge_label_test.cpu(), pred_test.cpu())

    if epoch % 5 == 0:
        print(f"Epoch {epoch:02d}, Loss {loss:.4f}, AUC {auc:.4f}, AP {ap:.4f}")

# ----------------------------
# 7. Inference on new unseen order
# ----------------------------
model.eval()
with torch.no_grad():
    # Add a new order node
    new_order_idx = num_orders
    new_order_x = torch.randn(1, 16).to(device)
    data['order'].x = torch.cat([data['order'].x, new_order_x], dim=0)

    # Connect to known products
    products_for_new_order = torch.tensor([0, 5, 7], device=device)
    new_edges = torch.stack([
        torch.full((len(products_for_new_order),), new_order_idx, device=device),
        products_for_new_order
    ])
    data['order', 'buys', 'product'].edge_index = torch.cat([data['order', 'buys', 'product'].edge_index, new_edges], dim=1)

    # Compute embeddings
    z_dict = model(data.x_dict, data.edge_index_dict)

    # Score all parts
    candidate_parts = torch.arange(num_parts, device=device)
    edge_label_index = torch.stack([
        torch.full((num_parts,), new_order_idx, device=device),
        candidate_parts
    ])
    scores = decoder(z_dict, edge_label_index).sigmoid()
    topk = torch.topk(scores, k=5)
    print("\nTop predicted parts for new order:", topk.indices.tolist())
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.data import HeteroData
from torch_geometric.nn import SAGEConv, HeteroConv
from torch_geometric.utils import negative_sampling
from sklearn.metrics import roc_auc_score, average_precision_score
import matplotlib.pyplot as plt

# ----------------------------
# 1. Synthetic heterogeneous graph
# ----------------------------
num_orders = 50
num_products = 20
num_parts = 30

data = HeteroData()
data['order'].x = torch.randn(num_orders, 16)
data['product'].x = torch.randn(num_products, 16)
data['part'].x = torch.randn(num_parts, 16)

# edges
data['order', 'buys', 'product'].edge_index = torch.stack([
    torch.randint(0, num_orders, (200,)),
    torch.randint(0, num_products, (200,))
], dim=0)

data['product', 'contains', 'part'].edge_index = torch.stack([
    torch.randint(0, num_products, (300,)),
    torch.randint(0, num_parts, (300,))
], dim=0)

edge_index = torch.stack([
    torch.randint(0, num_orders, (150,)),
    torch.randint(0, num_parts, (150,))
], dim=0)

# Train/test split
perm = torch.randperm(edge_index.size(1))
train_idx = perm[:int(0.8*edge_index.size(1))]
test_idx = perm[int(0.8*edge_index.size(1)):]
data['order', 'needs', 'part'].edge_index = edge_index[:, train_idx]
test_edges = edge_index[:, test_idx]

# reverse edges
data['product', 'rev_buys', 'order'].edge_index = data['order', 'buys', 'product'].edge_index.flip(0)
data['part', 'rev_contains', 'product'].edge_index = data['product', 'contains', 'part'].edge_index.flip(0)
data['part', 'rev_needs', 'order'].edge_index = data['order', 'needs', 'part'].edge_index.flip(0)

# ----------------------------
# 2. HeteroGNN encoder
# ----------------------------
class HeteroGNN(nn.Module):
    def __init__(self, hidden_channels):
        super().__init__()
        self.conv1 = HeteroConv({
            ('order','buys','product'): SAGEConv((-1,-1), hidden_channels),
            ('product','contains','part'): SAGEConv((-1,-1), hidden_channels),
            ('order','needs','part'): SAGEConv((-1,-1), hidden_channels),
            ('product','rev_buys','order'): SAGEConv((-1,-1), hidden_channels),
            ('part','rev_contains','product'): SAGEConv((-1,-1), hidden_channels),
            ('part','rev_needs','order'): SAGEConv((-1,-1), hidden_channels),
        }, aggr='mean')
        self.conv2 = HeteroConv({
            ('order','buys','product'): SAGEConv((-1,-1), hidden_channels),
            ('product','contains','part'): SAGEConv((-1,-1), hidden_channels),
            ('order','needs','part'): SAGEConv((-1,-1), hidden_channels),
            ('product','rev_buys','order'): SAGEConv((-1,-1), hidden_channels),
            ('part','rev_contains','product'): SAGEConv((-1,-1), hidden_channels),
            ('part','rev_needs','order'): SAGEConv((-1,-1), hidden_channels),
        }, aggr='mean')

    def forward(self, x_dict, edge_index_dict):
        x_dict = self.conv1(x_dict, edge_index_dict)
        x_dict = {k: F.relu(v) for k,v in x_dict.items()}
        x_dict = self.conv2(x_dict, edge_index_dict)
        return x_dict

# ----------------------------
# 3. Decoder
# ----------------------------
class DotProductDecoder(nn.Module):
    def forward(self, z_dict, edge_index):
        src, dst = edge_index
        return (z_dict['order'][src] * z_dict['part'][dst]).sum(dim=-1)

def get_link_labels(pos_edge_index, neg_edge_index):
    pos_y = torch.ones(pos_edge_index.size(1))
    neg_y = torch.zeros(neg_edge_index.size(1))
    return torch.cat([pos_y, neg_y], dim=0)

# ----------------------------
# 4. Model + optimizer
# ----------------------------
device = 'cuda' if torch.cuda.is_available() else 'cpu'
model = HeteroGNN(hidden_channels=32).to(device)
decoder = DotProductDecoder()
data = data.to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

# ----------------------------
# 5. Training loop with metric tracking
# ----------------------------
num_epochs = 50
train_losses = []
val_aucs = []
val_aps = []

for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()
    z_dict = model(data.x_dict, data.edge_index_dict)

    pos_edge_index = data['order','needs','part'].edge_index
    neg_edge_index = negative_sampling(
        edge_index=pos_edge_index,
        num_nodes=(num_orders,num_parts),
        num_neg_samples=pos_edge_index.size(1)
    )

    edge_label_index = torch.cat([pos_edge_index, neg_edge_index], dim=1)
    edge_label = get_link_labels(pos_edge_index, neg_edge_index).to(device)

    out = decoder(z_dict, edge_label_index)
    loss = F.binary_cross_entropy_with_logits(out, edge_label)
    loss.backward()
    optimizer.step()
    train_losses.append(loss.item())

    # Validation on test edges
    model.eval()
    with torch.no_grad():
        z_dict = model(data.x_dict, data.edge_index_dict)
        neg_test = negative_sampling(
            edge_index=data['order','needs','part'].edge_index,
            num_nodes=(num_orders,num_parts),
            num_neg_samples=test_edges.size(1)
        )
        edge_label_index_test = torch.cat([test_edges, neg_test], dim=1)
        edge_label_test = get_link_labels(test_edges, neg_test).to(device)
        pred_test = decoder(z_dict, edge_label_index_test).sigmoid()
        auc = roc_auc_score(edge_label_test.cpu(), pred_test.cpu())
        ap = average_precision_score(edge_label_test.cpu(), pred_test.cpu())
        val_aucs.append(auc)
        val_aps.append(ap)

    if epoch % 5 == 0:
        print(f"Epoch {epoch:02d}, Loss {loss:.4f}, Val AUC {auc:.4f}, Val AP {ap:.4f}")

# ----------------------------
# 6. Plot training loss and val metrics
# ----------------------------
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.plot(train_losses, label='Train Loss')
plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.title('Training Loss')
plt.grid(True); plt.legend()

plt.subplot(1,3,2)
plt.plot(val_aucs, label='Val ROC-AUC', color='orange')
plt.xlabel('Epoch'); plt.ylabel('AUC'); plt.title('Validation ROC-AUC')
plt.grid(True); plt.legend()

plt.subplot(1,3,3)
plt.plot(val_aps, label='Val AP', color='green')
plt.xlabel('Epoch'); plt.ylabel('AP'); plt.title('Validation Average Precision')
plt.grid(True); plt.legend()

plt.tight_layout()
plt.show()

# ----------------------------
# 7. Inference for new unseen order
# ----------------------------
model.eval()
with torch.no_grad():
    new_order_idx = num_orders
    new_order_x = torch.randn(1,16).to(device)
    data['order'].x = torch.cat([data['order'].x, new_order_x], dim=0)

    products_for_new_order = torch.tensor([0,5,7], device=device)
    new_edges = torch.stack([
        torch.full((len(products_for_new_order),), new_order_idx, device=device),
        products_for_new_order
    ])
    data['order','buys','product'].edge_index = torch.cat([data['order','buys','product'].edge_index, new_edges], dim=1)

    z_dict = model(data.x_dict, data.edge_index_dict)

    candidate_parts = torch.arange(num_parts, device=device)
    edge_label_index = torch.stack([
        torch.full((num_parts,), new_order_idx, device=device),
        candidate_parts
    ])
    scores = decoder(z_dict, edge_label_index).sigmoid()
    topk = torch.topk(scores, k=5)
    print("\nTop predicted parts for new order:", topk.indices.tolist())
```
